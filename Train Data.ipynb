```python
#To import the ultralytics library

from IPython import display
display.clear_output()
import ultralytics
ultralytics.checks()
```

```python
#To import YOLO class from Ultralytics module
from ultralytics import YOLO
 
# Load the model
model = YOLO('yolov8s.pt')
 
# Training
results = model.train(
   data='..\\data.yaml',
   imgsz=640,
   epochs=500,
   patience=30,
   lr0=0.001,
   plots=True,
   name='Final'
)
```

```python
from ultralytics import YOLO

model = YOLO("..\\Runs\\detect\\Final\\Weights\\best.pt")
result = model.predict(source="..\\Testing\\test.mp4", show=True)
```
